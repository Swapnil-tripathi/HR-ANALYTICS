**DATA VISUALIZATION:**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

**CORRELATION**

```python
#Correlation between all the columns
plt.figure(figsize=(12,10))#Set the figure size
corr_matrix=df.corr()#df.corr() is used to find correlation between all the columns of the dataset.
sns.heatmap(corr_matrix,annot=True,fmt=".1f",linewidth=.5,cmap='magma')#Heatmap in Seaborn library is used to plot rectangular data as color-encoded matrix.
#It is used to visualize 2D Data such as Correlation matrix.
plt.title('Correlation')
plt.show()
```

**OVERTIME**

```python
#Visualization of OverTime using barplot of Seaborn
plt.figure(figsize=(15,5))
sns.barplot(x=df['Department'],y=df['HourlyRate'],hue=df['OverTime'],width=0.3,palette='pastel')
plt.xlabel('Department')
plt.ylabel('Hourly_Rate')
plt.title('Hourly Rate by Department and OverTime')
plt.show()
```

```python
#Relationship between Overtime and age
sns.lineplot(x='OverTime',y='Age',data=df,color='pink')
plt.xlabel('OverTime_hours')
plt.ylabel('Age')
plt.title('Relationship between OverTime and Age')
plt.show()
```

**MARITAL STATUS**

```python
#Visualizing the Marital_Status
plt.figure(figsize=(15,5))
sns.barplot(data=df,x='Department',y='EmployeeNumber',hue='MaritalStatus',width=0.3,palette='Reds')
plt.xlabel('Department')
plt.ylabel('EmployeeNumber')
plt.title('EmployeeNumber by department and MaritalStatus')
```

```python
marital_status=df['MaritalStatus'].value_counts()
print(marital_status)
colors_=sns.color_palette('deep')
plt.pie(marital_status,labels=['Married','Single','Divorced'],colors=colors_,autopct='%.0f%%')
plt.title('Marital Status Count')
```

**JOB ROLE**

```python
#Visualizing the JobRole
plt.figure(figsize=(22,5))
sns.lineplot(data=df,x='JobRole',y='EmployeeNumber',marker='o',markersize=10,color='orange')
plt.title('EmployeeNumber by JobRole')
```

```python
jobrole_count=df['JobRole'].value_counts()
print(jobrole_count)
label_=['Sales Executive','Research Scientist','Laboratory Technician','Manufacturing Director','Healthcare Representa','Manager','Sales Representative','Research Director','Human Resources']
plt.pie(jobrole_count,labels=label_,autopct='%.0f%%')
plt.title('Jobrole Count')
```

**GENDER**

```python
#Visualizing Gender with the help of count plot
plt.figure(figsize=(22,5))
sns.countplot(data=df,x='JobRole',hue='Gender',palette='icefire')
plt.title('Job Role by Gender')
```

```python
Gender_count=df['Gender'].value_counts()
print(Gender_count)
palette_color=sns.color_palette('pastel')
plt.pie(Gender_count,labels=['Male','Female'],colors=palette_color,autopct='%.0f%%',explode=[0.1,0])
plt.title('Gender count')
plt.show()
```

**EDUCATION FIELD**

```python
#Visualizing Education field by comparing gender
plt.figure(figsize=(30,6))
sns.barplot(data=df,x='EducationField',y='MonthlyIncome',hue='Gender',palette='pastel')
plt.title('Monthly Income by Education field and Gender')
plt.show()
#From the above graph we can conclude that in LifeSciences,Medical and other monthly income of both male and female is approximately equal
#but in Marketing,Technical degree and Human Resources monthly income of female is greater than male.
```
